```python

# Finding the customer that brings in the highest profit to the sales company
customer_profit = df.groupby('Customer Name')['Profit'].sum()

customer_with_max_profit = customer_profit.idxmax()

# Getting the maximum profit value
max_profit = customer_profit.max()

# Printing the customer with the highest profit and the corresponding profit value
print(f"The customer with the highest profit is: {customer_with_max_profit}")
print(f"Their profit value is: {max_profit}")

```

```python
plt.scatter(df['Sales'], df['Profit'])
plt.xlabel('Sales')
plt.ylabel('Profit')
plt.title('Correlation between Sales and Profit')
plt.show()
```

```python
x = df['Sales']
y = df['Profit']
corr_matrix = np.corrcoef(x, y)
corr_coef = corr_matrix[0, 1]
m, b = np.polyfit(x, y, 1)

plt.scatter(x, y)
plt.plot(x, m*x + b, color='red')
plt.xlabel('Sales')
plt.ylabel('Profit')
plt.title(f'Correlation between Sales and Profit: {corr_coef:.2f}')
plt.show()
```
